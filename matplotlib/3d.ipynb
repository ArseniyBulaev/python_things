```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget
```

```python
fig = plt.figure(figsize=(7,4))
#ax_3d = Axes3D(fig)
ax_3d = plt.subplot(projection='3d')
x = np.linspace(0, 10, 50)
z = np.cos(x)
ax_3d.plot(x, x, z)
ax_3d.set_xlabel('x')
ax_3d.set_ylabel('y')
ax_3d.set_zlabel('z')
```

```python
x, y = np.arange(-2*np.pi, 2*np.pi, 0.2), np.arange(-2*np.pi, 2*np.pi, 0.2)
xgrid, ygrid = np.meshgrid(x, y)
fig = plt.figure(figsize=(7,4))
ax_3d = plt.subplot(projection='3d')
b = ax_3d.plot_surface(xgrid,
    ygrid,
    np.sin(xgrid) * np.sin(ygrid) / (xgrid*ygrid),
    cmap='magma',
    rstride=5,
    cstride=5,
)
plt.colorbar(b)
```

```python
fig = plt.figure(figsize=(7,4))
ax_3d = plt.subplot(projection='3d')
b = ax_3d.contourf(xgrid,
    ygrid,
    np.sin(xgrid) * np.sin(ygrid) / (xgrid*ygrid),
    cmap='magma',
    rstride=5,
    cstride=5,
)
```

```python
zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid*ygrid)
fig, ax = plt.subplots(1,2)
c1 = ax[0].contour(xgrid, ygrid, zgrid, [0.2])
ax[1].contourf(xgrid, ygrid, zgrid)
ax[0].clabel(c1)
```
