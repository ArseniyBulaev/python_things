```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
```

```python
# backend
#matplotlib.use("Qt5Agg")
print(matplotlib.get_backend())
```

```python
x = np.arange(0, 101)
y = x**2
lines = plt.plot(x, y, x, y**1.05, color=(0.7, 0.2, 0.3, 0.5))
print(lines)
plt.setp(lines[0], linestyle='--', color=(0,1,0))
plt.grid() # Grid
```

```python
#  Fill between
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y, color='purple', where =(y > 0), alpha = 0.5)
plt.fill_between(x, y, color='green', where =(y < 0), alpha = 0.5)
plt.grid()
```
