```python
import numpy as np
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,1,0.01)
```

```python
ws = [0.25, 0.75]
hs = [1,2,2]
fig = plt.figure()

gs = GridSpec(ncols=2, nrows=3, figure=fig, width_ratios=ws, height_ratios=hs)
ax1 = plt.subplot(gs[0, :])
ax1.plot(x, 2*x, color=(1,0,0))
ax2 = plt.subplot(gs[1, 0])
ax2.plot(x, 3*x, color=(0,1,0))
ax3 = plt.subplot(gs[1, 1])
ax3.plot(x, 4*x, color=(0,0,1))
ax4 = plt.subplot(gs[2, 1])
ax4.plot(x, 5*x, color=(1,1,0))
```
