```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('dog-shit-doodoo.png')
```

```python
cmap_names = [str(cmap) for cmap in plt.colormaps]
image_as_array = np.array(img)
fig, ax = plt.subplots(85,2)
fig.set_size_inches(10, 100)
for i in range(len(cmap_names)):
    ax[i//2, i%2].imshow(image_as_array, cmap=cmap_names[i])
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot()
b = ax.imshow(img)
plt.colorbar(b)
plt.show()
```
