```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import *
```

```python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
ax.plot(np.arange(0,5, 0.25), label='foo')
ax.plot(np.arange(0,10,0.5), label='bar')
ax.legend(loc='upper right')
```

```python
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
l1x = np.arange(-0.5,1.5, 0.00001)
l1 = Line2D(l1x, np.sqrt(1**2-(l1x - 0.5)**2) + 0.5)
l2 = Line2D(l1x, -np.sqrt(1**2-(l1x - 0.5)**2) + 0.5)
rect = Rectangle((0,0), 0.5, 0.5, facecolor='g')
ax.add_line(l1)
ax.add_line(l2)
ax.add_patch(rect)
ax.set_xlim(xmin=-0.5, xmax=1.5)
ax.set_ylim(ymin=-0.5, ymax=1.5)
```
