```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linm
fig = plt.figure(figsize=(7,4))
ax = fig.add_subplot()
ax.plot(np.arange(1, 5, 0.25))
ax.set(xlim=(-2,2), ylim=(-2,2))
ax.set_xlim(xmin=-5, xmax=10)
ax.set_xlim(xmin=-5)
```

```python
# Locator
from matplotlib.ticker import NullLocator, LinearLocator, MultipleLocator, IndexLocator
data = np.arange(1,5,0.25)
fig, ax = plt.subplots(2,2)
fig.set_size_inches(12,6)
# NullLocator
ax[0,0].plot(data)
ax[0,0].xaxis.set_major_locator(NullLocator())
# LinearLocator
ax[0,1].plot(data)
ax[0,1].xaxis.set_major_locator(LinearLocator(5))
# MultipleLocator
ax[1,0].plot(data)
ax[1,0].xaxis.set_major_locator(MultipleLocator(np.pi))
# IndexLocator
x = np.arange(-np.pi/2, np.pi, 0.1)
ax[1,1].plot(x, np.sin(x))
ax[1,1].xaxis.set_major_locator(IndexLocator(base=0.5, offset=1.57))
```

```python
# FixedLocator
from matplotlib.ticker import FixedLocator
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(data)
ax.xaxis.set_major_locator(FixedLocator([1,2]))
```

```python
# LogLocator
from matplotlib.ticker import LogLocator
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(data)
ax.xaxis.set_major_locator(LogLocator(base=2))
```

```python
# LogLocator
from matplotlib.ticker import MaxNLocator
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(data)
ax.minorticks_on()
ax.grid(which='major', lw=2)
ax.grid(which='minor')
ax.xaxis.set_major_locator(MaxNLocator(5))
```
