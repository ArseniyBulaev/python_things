```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
```

```python
y = np.arange(0, 5, 1)
x = np.array([a * a for a in y])
y2 = [0, 2, 3, 4, 5, 7]
x2 = [i + 1 for i in y2]
lines = plt.plot(x, y, x2, y2)
plt.minorticks_on()
plt.grid(which='minor', color='grey', alpha=0.5)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
#plt.figtext(0.3, 0.5, "Hello guys. I'm in NY City. Just hanging out")
ax.text(0.2, 0.5, "Hello guys. I'm in NY City. Just hanging out", va='center', color=(0,0,1))
plt.title("Wierd SHIT")
ax.set_xlabel("FOO")
ax.set_ylabel("BAR")
ax.set(facecolor='orange')
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
ax.annotate("Google it", xy=(0.5, 0.55), xytext=(0.6, 0.7),
    arrowprops={'facecolor':'gray', 'shrink':0.1}
)
```
