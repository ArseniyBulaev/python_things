```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.grid()
ax.set_xticklabels([])
ax.set_yticklabels([])
```

```python
from matplotlib.ticker import NullFormatter
fig = plt.figure()
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.grid()
ax.xaxis.set_major_formatter(NullFormatter())
ax.yaxis.set_major_formatter(NullFormatter())
```

```python
from matplotlib.ticker import FormatStrFormatter
fig = plt.figure()
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.grid()
ax.xaxis.set_major_formatter(FormatStrFormatter("blabla %.2f"))
ax.yaxis.set_major_formatter(FormatStrFormatter("%.2f"))
```

```python
def format0x(x, pos):
    return f"even {x}" if pos % 2 == 0 else f"odd {x}"
    


from matplotlib.ticker import FuncFormatter
fig = plt.figure()
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.grid()
ax.xaxis.set_major_formatter(FuncFormatter(format0x))
```

```python
from matplotlib.ticker import ScalarFormatter
fig = plt.figure()
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e5)
ax.grid()
sf = ScalarFormatter()
#sf.set_powerlimits((-4,4))
#sf.set_scientific(True)
ax.yaxis.set_major_formatter(sf)

# Global 
import matplotlib
matplotlib.rcParams["axes.formatter.limits"] = (-4,4)
```

```python
from matplotlib.ticker import FixedFormatter
fig = plt.figure()
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e5)
ax.grid()
sf = FixedFormatter(list("qwertyuiop"))
ax.yaxis.set_major_formatter(sf)
```
